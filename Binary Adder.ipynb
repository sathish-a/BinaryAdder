# Making a model which does binary addition 

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split 
```

<b>No.Of.Bits model can accept</b>

```python
bits = 8
```

<b> Generating numbers having 8 bits </b>

```python
def generate(v):
    x_ = []
    for i in range(254):
        y_ = []
        z = bin(i + v)[2:].zfill(bits)
        for j in z:
            y_.append(int(j))
        x_.append(y_)
    return np.array(x_)
```

<b>Creating Dataset</b> 

```python
def get_dataset():
    return generate(0), generate(1)
```

<b>Splitting the data in 70:30 ratio randomly </b>

```python
x,y = get_dataset()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
print (x_train.shape,y_train.shape)
print (x_train[1],y_train[1])
```

<b>Creating a model, which accepts 8 bit number and return an 8 bit number<b>

```python
model = Sequential()
model.add(Dense(units=128, input_dim=bits, activation='sigmoid'))
model.add(Dense(units=64, activation='sigmoid'))
model.add(Dense(units=bits, activation="sigmoid"))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=500, batch_size=5)
```

<b>Saving Model</b>

```python
model.save_weights("modeladd.h5")
```

<b>Loading the already existing model</b>

```python
model.load_weights("modeladd.h5")
```

<b>Evaluating the model with test data</b>

```python
model.evaluate(x_test,y_test,verbose=1,batch_size=128)
```
